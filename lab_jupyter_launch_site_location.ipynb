```python
import folium
from folium.plugins import MarkerCluster

# ベースマップ
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# 成功/失敗でレイヤー分割
fg_success = folium.FeatureGroup(name='Success (green)', show=True)
fg_failure = folium.FeatureGroup(name='Failure (red)', show=True)

# 各レイヤーにクラスタを配置
cl_success = MarkerCluster(name='Success Cluster').add_to(fg_success)
cl_failure = MarkerCluster(name='Failure Cluster').add_to(fg_failure)

# レコードを振り分けて追加
for _, r in spacex_df.iterrows():
    popup_txt = f"{r['Launch Site']}<br>Outcome: {'Success' if r['class']==1 else 'Failure'}"
    marker = folium.Marker(
        location=[r['Lat'], r['Long']],
        popup=popup_txt,
        icon=folium.Icon(color=r['marker_color'], icon='rocket')
    )
    if r['class'] == 1:
        cl_success.add_child(marker)
    else:
        cl_failure.add_child(marker)

# レイヤーをマップへ
fg_success.add_to(site_map)
fg_failure.add_to(site_map)

# レイヤーコントロール
folium.LayerControl(collapsed=False).add_to(site_map)

# 表示（Jupyterではセル末尾に置く）
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition

# マウスオーバーで座標を表示するプラグインを追加
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',        # 表示位置
    separator=' Long: ',        # 区切り文字
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# 表示
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
from folium.features import DivIcon

# Launch Site (例: CCAFS SLC-40)
launch_site_lat, launch_site_lon = 28.563197, -80.576820

# MousePositionで取得した「海岸線に最も近い点」の座標（例）
coastline_lat, coastline_lon = 28.56367, -80.57163

# 距離を計算
distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon,
    coastline_lat, coastline_lon
)

print(f"Distance to coastline: {distance_coastline:.2f} km")

# 海岸線にマーカーを追加（距離をラベル表示）
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % 
             "{:10.2f} KM".format(distance_coastline),
    )
)

site_map.add_child(distance_marker)

# 地図を表示
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from folium.features import DivIcon

def add_proximity_point(name, lat, lon, color='#1e90ff'):
    """
    name: 表示名（'Coastline', 'Railway', 'Highway', 'Nearest City' など）
    lat, lon: 近接地点の座標（MousePositionで取得）
    color: ライン色
    """
    # 距離(km)
    d_km = calculate_distance(launch_site_lat, launch_site_lon, lat, lon)

    # 距離ラベル付きマーカー
    folium.Marker(
        location=[lat, lon],
        popup=f"{name}<br>{d_km:.2f} km from launch site",
        icon=DivIcon(
            icon_size=(120, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;color:#d35400;"><b>{name}: {d_km:.2f} KM</b></div>'
        )
    ).add_to(site_map)

    # 発射場 ↔ 近接地点のライン
    folium.PolyLine(
        locations=[(launch_site_lat, launch_site_lon), (lat, lon)],
        color=color, weight=3, opacity=0.8,
        tooltip=f"{name} distance: {d_km:.2f} km"
    ).add_to(site_map)

    return d_km
```
